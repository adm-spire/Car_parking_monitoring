```python
import cv2
from ultralytics import YOLO

# Load YOLO model
model = YOLO("D:\\Downloads\\yolov8s_fold_0.pt")
#"D:\\Downloads\\yolov8s_fold_0.pt"
#"C:\\Users\\rauna\\runs\\detect\\train27\\weights\\best.pt"
# Open the video file
cap = cv2.VideoCapture("D:\Downloads\car_parking.mp4")

# Specify the desired frame rate (frames per second)
desired_fps = 15  # Adjust this value as needed

while cap.isOpened():
    ret, frame = cap.read()
    
    # Check if frame is read successfully
    if not ret:
        break
    
    # Detect objects in the frame
    results = model(frame, conf=0.5)

    for i, r in enumerate(results):
        # Plot the bounding boxes on the frame
        im_bgr = r.plot()  # BGR-order numpy array
        #im_rgb = cv2.cvtColor(im_bgr, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB
        
        # Display output video
        cv2.imshow('Results', im_bgr)
        
        # Save results to disk
        r.save(filename=f'results{i}.jpg')
        
        # Adjust the delay based on the desired frame rate
        delay = int(1000 / desired_fps)  # Delay in milliseconds
        if cv2.waitKey(delay) & 0xFF == ord('q'):
            break

# Release video capture and close all windows
cap.release()
cv2.destroyAllWindows()
```
