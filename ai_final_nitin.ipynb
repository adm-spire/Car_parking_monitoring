```python
import cv2
from ultralytics import YOLO

# Load your YOLO model
model = YOLO("D:\\Downloads\\yolov8s_fold_0.pt")


# Open video capture
cap = cv2.VideoCapture("D:\\Downloads\\car_parking.mp4")

# Set desired frame rate (frames per second)
frame_rate = 30  # Adjust as needed

while cap.isOpened():
    ret, frame = cap.read()

    if not ret:
        break

    # Apply your mask image processing here
    mask_img = cv2.imread("C:\\Users\\rauna\\Pictures\\ai_car_sample_2.png", cv2.IMREAD_GRAYSCALE)
    mask_img = cv2.resize(mask_img, (frame.shape[1], frame.shape[0]))
    masked_frame = cv2.bitwise_and(frame, frame, mask=mask_img)

    # Run inference on the masked frame
    
    results = model.predict(masked_frame, conf=0.3)
    

    for i, r in enumerate(results):
        im_bgr = r.plot()  # BGR-order numpy array
        #im_rgb = cv2.cvtColor(im_bgr, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB
        cv2.imshow('Results', im_bgr)  # Display output video
        r.save(filename=f'results{i}.jpg')  # Save results to disk

    # Adjust frame rate by introducing a delay
    delay = int(1000 / frame_rate)  # Delay in milliseconds
    if cv2.waitKey(delay) & 0xFF == ord('q'):
        break

# Release video capture and close OpenCV windows
cap.release()
cv2.destroyAllWindows()

```
