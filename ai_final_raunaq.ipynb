```python
import cv2
from ultralytics import YOLO
import numpy as np
import torch


img= cv2.imread("C:\\Users\\rauna\\Pictures\\Screenshots\\ai_car_sample.png")
model = YOLO('yolov8m-seg.pt')
results = model.predict(source=img.copy(), save=True, save_txt=False, stream=True)
for result in results:
    # get array results
    masks = result.masks.data
    boxes = result.boxes.data
    # extract classes
    clss = boxes[:, 5]
    # get indices of results where class is 3 (car in COCO)
    people_indices = torch.where(clss == 3)
    # use these indices to extract the relevant masks
    people_masks = masks[people_indices]
    # scale for visualizing results
    
    people_mask = torch.any(people_masks, dim=0).int() * 255
    # save to file
   
    cv2.imwrite(str(model.predictor.save_dir / 'merged_segs.jpg'),people_mask.cpu().numpy())
```

```python
import cv2
from ultralytics import YOLO

# Load your YOLO model
model = YOLO("C:\\Users\\rauna\\runs\\detect\\train8\\weights\\best.pt")



# Open video capture
cap = cv2.VideoCapture("D:\Downloads\car_parking.mp4")

while cap.isOpened():
    ret, frame = cap.read()

    # Load your mask image
    mask_img = cv2.imread("C:\\Users\\rauna\\runs\\segment\\predict3\\image0.jpg", cv2.IMREAD_GRAYSCALE)
    mask_img = cv2.resize(mask_img, (frame.shape[1], frame.shape[0]))
    
    if not ret:
        break
    
    # Apply the mask to the input frame
    masked_frame = cv2.bitwise_and(frame, frame, mask=mask_img)

    # Run inference on the masked frame
    results = model(masked_frame, conf=0.05)

    for i, r in enumerate(results):
        im_bgr = r.plot()  # BGR-order numpy array
        
        cv2.imshow('Results', im_bgr)  # Display output video
        
        # Save results to disk
        r.save(filename=f'results{i}.jpg')
        
        if cv2.waitKey(1) & 0xFF == ord('q'):
            cap.release()
            cv2.destroyAllWindows()
            break


```

```python
import cv2
import numpy as np
from ultralytics import YOLO

# Load your YOLO model
model = YOLO("yolov8m-seg.pt")

# Load your image
image = cv2.imread("C:\\Users\\rauna\\runs\\segment\\predict3\\image0.jpg")

# Run inference
results = model(image)

# Create a blank canvas with black background
mask = np.zeros_like(image, dtype=np.uint8)

# Loop through each detection result
for result in results:
    # Get segmentation mask and bounding box
    #mask_data = result.segmentation.data
    bbox = result.xyxy[0].cpu().numpy()

    # Draw the object on the mask
    cv2.fillPoly(mask, [np.array(bbox[:4], dtype=np.int32)], (255, 255, 255))

# Invert the mask
mask = cv2.bitwise_not(mask)

# Display or save the mask
cv2.imshow("Mask", mask)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
